#POS TAGGING USING CRF
An entity, or a part of text that is of interest would be of great use if it could be recognized, named and called to identify similar entities. A CRF is a sequence modeling algorithm which is used to identify entities or patterns in text, such as POS tags. This model not only assumes that features are dependent on each other, but also considers future observations while learning a pattern. In terms of performance, it is considered to be the best method for entity recognition.Since these models take into account previous data, we use features which are modelled from the data to feed into the CRF. These feature functions express certain characteristic of the sequence that the data point represents, such as the tag sequence **noun -> verb -> adjective**

#Splitting the data into train and test set

```python
train_set, test_set = train_test_split (tagged_sentence, test_size=0.2,random_state=1234)
print("Number of Sentences in Training Data ",len(train_set))
print("Number ef Sentences in Testing Data ",len(test_set))
```

#Defining the feature function

```python
def features(sentence, index):
  return {
      'is_first_capital':int(sentence[index][0].isupper()),
      'is_first_word': int(index==0),
      'is_last_word': int(index==len(sentence)-1),
      'is_complete_capital': int (sentence[index].upper()==sentence[index]),
			'prev_word':'' if index==0 else sentence[index-1],
      'next_word':'' if index==len(sentence)-1 else sentence[index+1],
      'is_numeric':int(sentence[index].isdigit()),
      'is_alphanumeric':
      int(bool((re.match('^(?=.*[0-9]$)(?=.*[a-zA-Z])',sentence[index] )))),
			'prefix_1':sentence[index][0],
			'prefix_2': sentence[index][:2],
			'prefix_3': sentence[index][:3],
			'prefix_4':sentence[index][:4],
			'suffix_1':sentence[index][-1],
			'suffix_2':sentence[index][-2:],
			'suffix_3':sentence[index][-3:],
			'suffix_4': sentence[index][-4:],
			'word_has_hyphen': 1 if '-' in sentence[index] else 0
}  
```

#Separate label and the sentences in both training and test data

#EXPLANATION
**untag(sentence):** This function takes a sentence that is a list of word-tag pairs and returns a list of words by stripping off the tags. It uses a list comprehension to iterate over each word-tag pair in the sentence and return only the word.

**prepareData(tagged_sentences):** This function takes a list of tagged sentences and prepares the data for training and testing the CRF model. 

**X_train, y_train = prepareData(train_set):** This line prepares the training data by calling the prepareData() function on the train_set variable, which is a list of tagged sentences.

X_test, y_test = prepareData(test_set): **bold text** This line prepares the testing data by calling the prepareData() function on the test_set variable, which is also a list of tagged sentences.

**print('X Training Data',X_train[0]):** This line prints the first example of the training data, which is a list of feature lists for the words in the first sentence.

**print('Y Training Data',y_train[0]):** This line prints the first example of the training labels, which is a list of tags for the words in the first sentence.

The dependencies for this code include the** features() function (not shown)**, which is used to extract features for each word in the sentences, and the train_set and test_set variables, which are lists of tagged sentences.


```python
def untag(sentence):
  return [word for word, tag in sentence]
def prepareData (tagged_sentences):
  X,y=[],[]
  for sentences in tagged_sentences:
    X.append([features(untag(sentences),index) for index in range(len(sentences))])
    y.append([tag for word, tag in sentences])
  return X,y

X_train,y_train=prepareData(train_set)

X_test, y_test=prepareData (test_set)

print('X Training Data',X_train[0])

print('Y Training Data',y_train[0])
```

#Fit the CRF model with the default parameter

#EXPLANATION
**crf = CRF(algorithm= 'lbfgs', c1=0.01, c2=0.1, max_iterations=100, all_possible_transitions=True):** This line creates an instance of the CRF class from the sklearn_crfsuite module with the specified hyperparameters. The 'lbfgs' algorithm is used to optimize the loss function, c1 and c2 are L1 and L2 regularization parameters, and max_iterations is the maximum number of iterations to run during training.
**all_possible_transitions=True** allows the CRF to consider all possible transitions between tags during training.

**crf.fit(X_train, y_train):** This line trains the CRF model on the training data X_train and its corresponding labels y_train.

**y_pred_train=crf.predict (X_train):** This line generates predicted labels for the training data using the trained CRF model.

**print("F1 Score on Train Data:", metrics.flat_f1_score(y_train,y_pred_train, average = "weighted", labels = crf.classes_)):** This line computes the F1 score on the training data using the flat_f1_score function from the sklearn_crfsuite.metrics module.
y_pred=crf.predict(X_test): This line generates predicted labels for the test data using the trained CRF model.




```python
crf = CRF(
    algorithm= 'lbfgs',
    c1=0.01,
    c2=0.1,

	max_iterations=100,
	all_possible_transitions=True
)

crf.fit(X_train, y_train)
 
y_pred_train=crf.predict (X_train)

print("F1 Score on Train Data:", metrics.flat_f1_score(y_train,y_pred_train, average = "weighted", labels = crf.classes_))
y_pred=crf.predict(X_test)
print("Fl Score on Test Data: ",metrics.flat_f1_score(y_test, y_pred, average ="weighted", labels = crf.classes_ ))
```

#Import features use to identify different POS Tags

```python
print('Score on ytest and ypred:',metrics.flat_accuracy_score(y_test, y_pred))
print('Score on ytest and ypred:',metrics.flat_accuracy_score(y_train,y_pred_train))
```
